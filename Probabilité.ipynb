## I) Contexte historique
Elles apparaissent dans des travaux de Bayes et de Moivre, écrits en anglais au XVIIIe siècle, même si c’est Laplace qui en a élaboré la notion. Les questions traitées par ces auteurs peuvent parfois surprendre (exemple : quelle est la probabilité que le soleil se lève demain, sachant qu'il s'est levé depuis le commencement du monde ?) ; néanmoins, les probabilités conditionnelles sont omniprésentes dans la vie courante et leur utilisation inappropriée mène facilement à de fausses affirmations.  
L’histoire des probabilités contribue à la réflexion sur la codification d'une théorie scientifique. On peut considérer que les origines du « calcul des probabilités » remontent au XVIIe siècle. Pascal, Huygens, Moivre, Bernoulli, Euler, d'Alembert appliquent les notions de variable aléatoire et d'espérance à des problèmes issus de questions liées aux jeux, aux assurances et à l’astronomie.  
Dans notre cas, nous tenterons d'approcher la valeur de $\pi$ à partir de cette notion de probabilité.

## II) Rappels et vocabulaire

**Exéprience aléatoire:**   
Protocole précis dont on peut prévoir l'issue mais qui peut être vérfiée.  

**Univers:**   
Ensemble des issues possibles d'une expérience aléatoire. On la note $\Omega$.   

**Evénement:**  
Sous-ensemble de l'univers $\Omega$.  

**Evénement élémentaire:**   
Evénement qui ne contient qu'un seul élément. On le note $e_i$.  

**Evénement certain:**  
C'est l'Univers $\Omega$.  

**Evénement impossible:**
C'est l'ensemble vide $\emptyset$.   

**Evénement contraire:**  
L'événement contraire de l'événement $A$, noté $\overline{A}$ est composé de tous les autres éléments de $\Omega$ qui ne sont pas dans $A$.

**Intersection de deux événements:**  
On appelle l'intersection de 2 événements $A$ et $B$, l'événement noté $A \cap B$ composé des éléments de $\Omega$ qui appartiennent à $A$ et à $B$.  
Schéma:  
  
  
  **Union de deux événements:**  
  On appelle l'union de 2 événements $A$ et $B$, l'événement noté $A \cup B$ composé des éléments de $\Omega$ qui appartiennent à $A$ ou à $B$. Il faut entendre ici **ou bien**, il s'agit d'un *ou exclusif*.  
  
  **Lois de Augustins De Morgan:**  
  
  $\overline{A \cup B} = \overline{A} \cap \overline{B}$  
  et    
  $\overline{A \cap B} = \overline{A} \cup \overline{B}$


## III) Probabilités conditionnelles

### 1) Définition  
<div class="alert alert-block alert-danger">

Lorsque $P(A)\ne 0$, on note alors $P_A(B)$ la probabilité d'avoir $B$ sachant que $A$ est réalisé. On a alors:   
$$P_A(B)=\frac{P(A \cap B)}{P(A)}$$  

</div> 


*Exemple:*  
Une urne contient 5 boules: 3 blanches et 2 noires. On effectue deux tirages consécutifs et sans remise. 
La probabilité de tirer une boule blanche au premier tirage est facile à estimer. 
Mais la probabilité de tirer une boule blanche au deuxième tirage dépend-elle de notre premier tirage?

### 3) Probabilité totale  

<div class="alert alert-block alert-danger">

Soit $A_1,A_2,...,A_n$ une partition de l'univers $\Omega$ alors on a pour tout événement $B$:  
$$P(B)=P(A_1 \cap B) +P(A_2 \cap B) + ... + P(A_n \cap B).$$  

</div>

 



## IV) Indépendance  
Reprenons le tout premier exemple de l'urne avec 5 boules: 3 blanches et 2 noires. Nous changeons une seule régle: on pioche deux fois de manère successive mais **avec remise**.  
Nous pouvons toujours calculer la probabilité de piocher une boule blanche au premier tirage. Qu'arrive-t-il pour le second tirage? 

On dit alors que le deuxième tirage est indépendant du premier.   


<div class="alert alert-block alert-danger">

On dit que deux événements $A$ et $B$  sont indépendants si et seulement si, $$P(A \cap B) = P(A) \times P(B)$$  

Ce qui revient à dire que $P(B)=P_A(B)=P_{\overline{A}}(B)$  

</div>

Cette relation nous apporte beaucoup car en vérifiant cette égalité, il est maintenant possible de prouver par un calcul que deux événements sont indépendants.
